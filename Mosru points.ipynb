```python
import pandas as pd
import re
```

You need to use Excel files from data.mos.ru

```python
input_path = input('Please, enter path to Excel file from data.mos.ru: ')
output_path = input('Please, enter path and name to output csv file: ')
df = pd.read_excel(input_path)
```

This code creates csv file with stops which will be used in ArcGIS

```python
routes = df["RouteNumbers"].tolist()
normal_routes = []
for elem in routes:
    strk = ''.join(elem)
    newelem = strk.split('; ')
    for i in newelem:
        normal_routes.append(i)
setted = set(normal_routes)
latitude = df["Latitude_WGS84"].tolist()
longitude = df["Longitude_WGS84"].tolist()
for lat in latitude:
    latstr = str(lat)
    ff = re.findall(r'^\d\d', latstr)
    ffstr = str(ff[0])
    fint = int(ffstr)
    if fint == 37:
        w_lat = lat
        indx = latitude.index(w_lat)
        w_lon = longitude[indx]
        latitude.pop(indx)
        latitude.insert(indx, w_lon)
        longitude.pop(indx)
        longitude.insert(indx, w_lat)
    else:
        pass
    numlist = []
latlist = []
lonlist = []
routelist = []
for x in setted:
    k = 0
    for elems in routes:
        newline = ''.join(elems)
        newlst = newline.split('; ')
        idx = routes.index(elems)
        for y in newlst:
            if x == y:
                numlist.append(k)
                k += 1
                latlist.append(latitude[idx])
                lonlist.append(longitude[idx])
                routelist.append(x)
            else:
                pass
string1 = ' '.join(routelist)
res = set(re.findall('\w', string1))
routlist_ENG =[]
for i in routelist:
    w = ''.join(i)
    reZ = w.replace(r'З','Z')
    reS = reZ.replace(r'С','S')
    reT = reS.replace(r'Т','T')
    reP = reT.replace(r'П','P')
    reK = reP.replace(r'К','K')
    reM = reK.replace(r'М','M')
    rec = reM.replace(r'с','c')
    reCe = rec.replace(r'Ц','Ce')
    rech = reCe.replace(r'ч','ch')
    rek = rech.replace(r'к','korot')
    reL = rek.replace(r'Л','L')
    reSh = reL.replace(r'Ш','Sh')
    reO = reSh.replace(r'О','O')
    reD = reO.replace(r'Д','D')
    reN = reD.replace(r'Н','N')
    rem = reN.replace(r'м','m')
    reE = rem.replace(r'Э','E')
    reb = reE.replace(r'б','b')
    reB = reb.replace(r'Б','B')
    reA = reB.replace(r'А','A')
    reV = reA.replace(r'В','V')
    ren = reV.replace(r'н','n')
    reye = ren.replace(r'е','ye')
    reee = reye.replace(r'ы','ee')
    rer = reee.replace(r'р','r')
    rece = rer.replace(r'ц','ce')
    resh = rece.replace(r'ш','sh')
    ref = resh.replace(r'ф','f')
    reo = ref.replace(r'о','o')
    reu = reo.replace(r'у','u')
    reF = reu.replace(r'Ф','F')
    rel = reF.replace(r'л','l')
    rea = rel.replace(r'а','a')
    red = rea.replace(r'д','d')
    rez = red.replace(r'з','z')
    rep = rez.replace(r'п','p')
    reG = rep.replace(r'Г','G')
    reya = reG.replace(r'я','ya')
    reeee = reya.replace(r'э','eee')
    reg = reeee.replace(r'г','g')
    ret = reg.replace(r'т','t')
    rev = ret.replace(r'в','v')
    rei = rev.replace(r'и','i')
    reyu = rei.replace(r'ю','yu')
    reMyagZnak = reyu.replace(r'ь','MyagZnak')
    reyo = reMyagZnak.replace(r'ё','yo')
    reZh = reyo.replace(r'Ж','Zh')
    reCh = reZh.replace(r'Ч','Ch')
    newline = rev.replace(r'х','x')
    routlist_ENG.append(newline)
transport = pd.DataFrame({'Stop_num':numlist, 'Route_num': routlist_ENG, 'X_coord': lonlist, 'Y_coord': latlist})
rn = transport['Route_num'].tolist()
rnset = set(rn)
Globalx = []
Globaly = []
Globalnum = []
for r in rnset:
    route_n = transport.loc[transport['Route_num'] == r]
    xlist = route_n['X_coord'].tolist()
    ylist = route_n['Y_coord'].tolist()
    localx = []
    localy =[]
    localnum = []
    for x in xlist:
        if x not in localx:
            localx.append(x)
    for y in ylist:
        if y not in localy:
            localy.append(y)
    for i in range(len(localx)):
        localnum.append(r)
    if len(localx) == len(localy):
        if len(localx)>1:
            for lx in localx:
                Globalx.append(lx)
            for ly in localy:
                Globaly.append(ly)
            for ln in localnum:
                Globalnum.append(ln)
        else:
            pass
    else:
        pass
normal_stops = pd.DataFrame({'Route_num': Globalnum, 'X_coord': Globalx,
                             'Y_coord': Globaly})
with open(output_path, "w") as f:
    normal_stops.to_csv(f, encoding ='cp1251')
```
